```python
# Import dependencies.
import matplotlib.pyplot as plt
# this is equivalent to from matplotlib import pyplot
# pyplot module will allow us to create all the necessary charts
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = "Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker ="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create a chart using Object-Oriented Functions
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, linewidth =2, label = "Boston", marker ="*", color="green")
ax.set_ylim(0,45)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.legend()
```

# Scatter Plot Creation using MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'x')
```

```python
plt.scatter(y_axis, x_axis, c="r", label = "Chicago")
plt.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

plt.gca().invert_yaxis()
```

## Create a Bubble Chart

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# input the values to new set with higher value 
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Refactor the code: instead of for loop, use list comprehension
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

## Create a Scatter Plot Using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

## Create a Bubble Chart Using the Object-Oriented Interface

```python
fig, ax= plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Skill Drill
fig, ax= plt.subplots()
ax.scatter(y_axis, x_axis, c = "skyblue", edgecolors = 'k', alpha=0.8,
           linewidth=2,
           s=[i *5 for i in y_axis], label = "Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0, 50)

plt.gca().invert_yaxis()
```

## Create a Pie Chart Using the MATLAB Method

```python
# Recall that #plt.bar, plt.barh, plt.scatter
# now we use plt.pie()
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Use an array data points for each value in th epie chart to explode
# 0.2 is there for the seventh month, July
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# To add the percentage of each wedge on the pie chart, we use the `autopict` parameter.
# also, format of one decimal place using `.1f%`.
plt.pie(y_axis, explode=explode_values, 
        labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Visualize the pie chart bigger for the better look
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, 
        labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')
plt.show()
```

## Create a Pie Chart Using the Object-Oriented Int.

```python
fig, ax= plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# Skill Drill,
# note that we insert figsize parameter in this code here
fig, ax= plt.subplots(figsize=(8,8))

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Input Startangle and Counterclock parameter to adjust the angle/order
ax.pie(y_axis,labels=x_axis,
      colors = colors,
      explode = explode_values,
      autopct = "%.1f%%",
      counterclock = False,
      startangle =90)
plt.show()
```
